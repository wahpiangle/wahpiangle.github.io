```python
#1a)
class Queue(object):
    #class initiation
    def __init__(self):
        self.queue = []
        
    #enqueue adds the value to the end of the queue
    def enqueue(self, value):
        self.queue.append(value)
        return self.queue
    
    #dequeue only removes the first element of the array when it is not empty
    def dequeue(self):
        if not self.queue:
            return None
        return self.queue.pop(0)
    
    def getQueue(self):
        return self.queue

    def size(self):
        return len(self.queue)


class Stack(object):
    def __init__(self):
        self.queue = []

    def enqueue(self, value):
        self.queue.append(value)
        return self.queue

    #dequeue only removes the last element of the array when it is not empty
    def dequeue(self):
        if not self.queue:
            return None
        return self.queue.pop()

    def getQueue(self):
        return self.queue

    def size(self):
        return len(self.queue)
```

```python
#1b)
class FIFOQueue(object):
    def __init__(self):
        self.queue = []

    def enqueue(self, value):
        self.queue = self.queue + [value]
        return self.queue

    def dequeue(self):
        if not self.queue:
            return None
        dequeued = self.queue[0]
        self.queue = self.queue[1:]
        return dequeued

    def getQueue(self):
        return self.queue

    def size(self):
        return len(self.queue)

class LIFOQueue(object):
    def __init__(self):
        self.queue = []

    def enqueue(self, value):
        self.queue = [value] + self.queue
        return self.queue

    def dequeue(self):
        if not self.queue:
            return None
        return self.queue.pop(0)

    def getQueue(self):
        return self.queue

    def size(self):
        return len(self.queue)
    
# Creating the FIFO and LIFO queues
fifo_queue = FIFOQueue()
lifo_queue = LIFOQueue()


fifo_queue.enqueue("apple")
fifo_queue.enqueue("banana")
fifo_queue.enqueue("cherry")
print("FIFO Queue Size:", fifo_queue.size())
print("FIFO Queue:", fifo_queue.getQueue())
print("Dequeued item:", fifo_queue.dequeue())
print("FIFO Queue after dequeue:", fifo_queue.getQueue())

#LIFO
lifo_queue.enqueue("apple")
lifo_queue.enqueue("banana")
lifo_queue.enqueue("cherry")
print("\nLIFO Queue Size:", lifo_queue.size())
print("LIFO Queue:", lifo_queue.getQueue())
print("Dequeued item:", lifo_queue.dequeue())
print("LIFO Queue after dequeue:", lifo_queue.getQueue())
```
